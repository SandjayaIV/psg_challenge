```python
file_path = "entrainement_xml.xml"
file = os.path.join(file_path)
```

```python
tree = ET.parse(file)
root = tree.getroot()
```

```python
root.attrib
```

```python
df_columns = ['groupe','membre']
df = pd.DataFrame(columns=df_columns)
for child in root:
    for child1 in child:
        df = df.append(pd.Series([child.get('id'),child1.get('id')],index=df.columns),ignore_index=True)
df.set_index(['groupe','membre'],inplace=True)
df.sort_index(inplace=True)
df
        
```

```python

df = pd.DataFrame(columns=root[0].keys())
for i in range(len([elem.tag for elem in root.iter('groupe')])):
    ls = list(root[i].attrib.values())
    df = df.append(pd.Series([ls[i] for i in range(len(ls))],index=root[0].keys()),ignore_index=True)
df
```

```python
df1 = pd.DataFrame(columns=root[0].keys())
df2 = pd.DataFrame(columns=root[0][0].keys())
for i in range(len([elem.tag for elem in root.iter('groupe')])):
    for j in range(len([elem.tag for elem in root[0].iter('membre')])):
        ls1 = list(root[i].attrib.values())
        print(ls1)
        ls2 = list(root[i][j].attrib.values())
        print(ls2)
        df1 = df1.append(pd.Series([ls1[i] for i in range(len(ls1))],index=root[0].keys()),ignore_index=True)
        df2 = df2.append(pd.Series([ls2[i] for i in range(len(ls2))],index=root[0][0].keys()),ignore_index=True)
df2
```

```python
df_f = pd.concat([df1, df2], axis=1, keys=['groupe', 'membre'],sort=True)
df_f
```
